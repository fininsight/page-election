```
import json # import json module

# with statement
with open('정세균.json') as json_file:
    json_data = json.load(json_file)
    print(json_data[0])
```

```
content_list = [x['content'] for x in json_data]
content_list[:5] 
```

```
df = pd.DataFrame({})
df['content'] = content_list
df.head()
```

```
! pip install tqdm
```

```
import re

def space_preprocessing(text):
    # 개행문자 제거
    text = re.sub(r'[\n\r\t]', '', str(text))
    return text

def special_preprocessing(text):
    # 개행문자 제거
    text = re.sub('[^ ㄱ-ㅣ가-힣 0-9 A-Z a-z]+', '', str(text))
    return text

df['content_clean'] = df['content'].apply(space_preprocessing).apply(special_preprocessing).apply(lambda x : x.strip())
df.head()
```

```
! pip install tqdm -U
```

```
from tqdm.notebook import tqdm
from gluonnlp.data import SentencepieceTokenizer
from kobert.utils import get_tokenizer
tok_path = get_tokenizer()
sp  = SentencepieceTokenizer(tok_path)

token_list = []

for content in tqdm(df['content_clean']):
    token_list.append(sp(content))

df['token_list'] = token_list
df.head()
```

```
df.to_csv('ko_bert_정세균.csv', index=False)
```
