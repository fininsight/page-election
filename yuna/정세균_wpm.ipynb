```python
from BytePairEncoder import BytePairEncoder
```

```python
n_iters = 5000
encoder = BytePairEncoder(n_iters)
encoder.train(content_list)
```

```python
df = pd.DataFrame({})

df['content_list'] = content_list
df.head()
```

```python
import re
def space_preprocessing(text):
    # 개행문자 제거
    text = re.sub(r'[\n\r\t]', '', str(text))
    return text

def special_preprocessing(text):
    # 개행문자 제거
    text = re.sub('[^ ㄱ-ㅣ가-힣 0-9 A-Z a-z]+', '', str(text))
    return text
```

```python
df['content_list_clean'] = df['content_list'].apply(space_preprocessing).apply(special_preprocessing)
df.head()
```

```python
from tqdm.notebook import tqdm

token_list = []

for content in tqdm(df['content_list_clean']):
    token_list.append(encoder.tokenize(content))

df['token_list'] = token_list
df.head()
```

```python
df.to_csv('정세균_wpm.csv', index=False)
```
