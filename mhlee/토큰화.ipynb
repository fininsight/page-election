```python
import pickle
from collections import defaultdict

def generate_dict_by_category(modelpath = "../../model/" ,min_count=100, max_count = 10000, categories=['정치', '경제']) :#, '경제', '사회', 'IT과학', '생활문화'])
    ret = defaultdict()
    for c in categories :
        pickle_in = open('tmp_ngram_{}.pickle'.format(c),"rb")
        ngram_freq = pickle.load(pickle_in)
        
        print(c)
        dic = defaultdict()
        for ngram in ngram_freq :
            if ngram[0] not in dic.keys() :
                dic[ngram[0]] = list(ngram)
            else :
                dic[ngram[0]][2] = dic[ngram[0]][2] + ngram[2]

        
        ret[c] = [dic[n] for n in dic.keys() if (dic[n][2] > min_count) & (dic[n][2] < max_count)]
        
    pickle_out = open("{}comps_dict.pickle".format(modelpath),"wb")
    pickle.dump(ret, pickle_out)
    pickle_out.close()
        
    return ret
        
a = generate_dict_by_category()        
        
```

```python
import pickle
from konlpy.tag import Mecab

def tokenize(content, modelpath="../../model/", morphs_file="morphs_dict", comp_file="comps_dict", category="정치") :
    def subtokenize(pos_list, dct) :
        pos_str = "[" + ", ".join([str(t) for t in pos_list]) + "]"
        for pattern in dct :
            src = ", ".join([str(i) for i in pattern[1]])
            tgt = str(pattern[0])
            pos_str = pos_str.replace(src, tgt)

        tokenized_text = eval(pos_str)
        return tokenized_text
    
    mecab = Mecab()
    tokenized_text = mecab.pos(content)
        
    morphs_dict = pickle.load(open('{}{}.pickle'.format(modelpath, morphs_file),"rb"))
    tokenized_text = subtokenize(tokenized_text, morphs_dict[category])
    
    comp_dict = pickle.load(open('{}{}.pickle'.format(modelpath, comp_file),"rb"))
    tokenized_text = subtokenize(tokenized_text, comp_dict[category])
    
    return tokenized_text    
```
