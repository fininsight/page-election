```python
import pandas as pd
startdate = "2020-01-01"
enddate = "2020-01-04"
```

```python
li_df = []

for d in pd.date_range(startdate, enddate) :
    li_df.append(pd.read_json("{}.json".format(d.strftime('%Y%m%d'))))
    
df = pd.concat(li_df)
```

```python
print(df.media.unique())
print(df.category1.unique())
print(df.category2.unique())
```
