```python
import pandas as pd
startdate = "2020-01-01"
enddate = "2020-01-05"
```

```python
li_df = []

for d in pd.date_range(startdate, enddate) :
    li_df.append(pd.read_json("{}.json".format(d.strftime('%Y%m%d'))))
    
df = pd.concat(li_df, sort=False)
```

```python
print(df.media.unique())
print(df.category1.unique())
print(df.category2.unique())
```

```python
print(df[df.category1=="생활/문화"].shape)
print(df[df.category1=="사회"].shape)
print(df[df.category1=="IT/과학"].shape)
print(df[df.category1=="세계"].shape)
print(df[df.category1=="경제"].shape)
print(df[df.category1=="정치"].shape)
```

```python
df_count = df.groupby([df['category1'],df['category2']])['category2'].agg(['count'])
df_count
```

```python
from datetime import timedelta, date,datetime
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days+1)):
        yield start_date + timedelta(n)

start_date = datetime.strptime('2020-03-01', "%Y-%m-%d")#date(2013, 1, 1)
end_date = datetime.strptime('2020-03-31',"%Y-%m-%d")#date(2015, 6, 2)
for single_date in daterange(start_date, end_date):
    print(single_date.strftime("%Y-%m-%d"))
```
